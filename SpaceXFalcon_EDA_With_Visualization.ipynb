```python
# pandas is a software library written for the Python programming language for data manipulation and analysis.
import pandas as pd
#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
```

```python
import pandas as pd

# Directly read the CSV from URL
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

df = pd.read_csv(URL)
df.head(5)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

### Visualize the relationship between Flight Number and Launch Site

```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
sns.catplot(
    x="FlightNumber",
    y="LaunchSite",
    hue="Class",
    data=df,
    kind="strip",  
    height=6,
    aspect=2
)

plt.title("Flight Number vs Launch Site colored by Class")
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")

plt.show()
```

### Visualize the relationship between Payload Mass and Launch Site

```python
# Plot a scatter point chart with x axis to be Pay Load Mass (kg) and y axis to be the launch site, and hue to be the class value
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="PayloadMass",
    y="LaunchSite",
    hue="Class",
    data=df,
    palette="Set1"
)

plt.title("Payload Mass vs Launch Site colored by Class")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Launch Site")
plt.grid(True)
plt.show()
```

### Visualize the relationship between success rate of each orbit type

```python
# Group by Orbit and calculate success rate
orbit_success_rate = df.groupby('Orbit')['Class'].mean().reset_index()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(
    x='Orbit',
    y='Class',
    data=orbit_success_rate,
    palette='viridis'
)

plt.title("Success Rate by Orbit")
plt.ylabel("Success Rate")
plt.xlabel("Orbit")
plt.ylim(0, 1)  # Because success rate is between 0 and 1
plt.grid(axis='y')
plt.show()
```

### Visualize the relationship between FlightNumber and Orbit type

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="FlightNumber",
    y="Orbit",
    hue="Class",
    data=df,
    palette="Set1"
)

plt.title("Flight Number vs Orbit colored by Class")
plt.xlabel("Flight Number")
plt.ylabel("Orbit")
plt.grid(True)
plt.show()
```

### Visualize the relationship between Payload Mass and Orbit type

```python
# Plot a scatter point chart with x axis to be Payload Mass and y axis to be the Orbit, and hue to be the class value
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="PayloadMass",
    y="Orbit",
    hue="Class",
    data=df,
    palette="Set1"
)

plt.title("Payload Mass vs Orbit colored by Class")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Orbit")
plt.grid(True)
plt.show()
```

### Visualize the launch success yearly trend

```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Year'] = year
df.head()
    
```

### Plot a line chart with x axis to be the extracted year and y axis to be the success rate. Compute average success rate per year

```python
yearly_success = df.groupby('Year')['Class'].mean().reset_index()

# Convert Year to integer for proper sorting
yearly_success['Year'] = yearly_success['Year'].astype(int)

# Sort by Year
yearly_success = yearly_success.sort_values('Year')

# Plot line chart
plt.figure(figsize=(10,6))
sns.lineplot(
    x='Year',
    y='Class',
    data=yearly_success,
    marker='o'
)

plt.title("Launch Success Rate by Year")
plt.xlabel("Year")
plt.ylabel("Average Success Rate")
plt.ylim(0,1)
plt.grid(True)
plt.show()
```

### Feature Engineering

```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### Create dummy variables to categorical columns

```python
# Applying One-Hot Encoding to selected columns
features_one_hot = pd.get_dummies(
    features,
    columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial']
)

# Display the first rows
features_one_hot.head()
```

### Cast all numeric columns to float64
